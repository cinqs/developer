# NPP-VIIRS夜光月度合成产品简介与Notebook示例
美国新一代国家极轨卫星(Suomi-NPP)搭载的可见光近红外成像辐射(VIIRS)传感器获取到的夜间灯光数据。夜间灯光影像中亮值斑块即为灯光区，使城镇等发光体明显区别于黑暗的无灯光背景区，避免了传统遥感影像城镇区域与非城镇区域混淆问题，在宏观尺度的城市研究中具有巨大的潜力和应用前景。此数据集为月度合成产品。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('NOAA_VIIRS_DNB_MONTHLY_V1_VCMSLCFG') \
             .filterDate("2021-01-01", "2021-10-31")

imgs = dataset.select(['average']);

map = aie.Map(
    center=[100.746,36.529],
    height=800,
    zoom=2
)

vis_params = {
    'bands': 'average',
    'min': 0.0,
    'max': 60.0
}

map.addLayer(
    imgs,
    vis_params,
    'Nighttime average',
    bounds=imgs.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('NOAA_VIIRS_DNB_MONTHLY_V1_VCMSLCFG_20150201').select(['average'])

map = aie.Map(
    center=[100.746,36.529],
    height=800,
    zoom=2
)

vis_params = {
    'bands': 'average',
    'min': 0.0,
    'max': 60.0
}

map.addLayer(
    img,
    vis_params,
    'Nighttime average',
    bounds=img.getBounds()
)
map
```
