# ALOS Global Digital Surface Model (AW3D30)简介与Notebook示例

该数据集是由全色遥感立体测绘仪（PRISM）制作的全球数字地表模型（DSM），水平分辨率约为30米（约为1角秒），该仪器是高级陆地观测卫星 "ALOS" 上的光学传感器。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定需要检索的区域
feature_collection = aie.FeatureCollection('China_Province') \
                        .filter(aie.Filter.eq('province', '浙江省'))

geometry = feature_collection.geometry()

# 指定检索数据集，可设置检索的空间范围
dataset = aie.ImageCollection('JAXA_ALOS_AW3D30_V3_2') \
             .filterBounds(geometry) \
             .limit(10)

imgs = dataset.select(['DSM'])

map = aie.Map(
    center=feature_collection.getCenter(),
    height=800,
    zoom=7
)

vis_params = {
    'bands': 'DSM',
    'min': 0,
    'max': 100,
    'palette': [
        '#0000ff', '#00ffff', '#ffff00', '#ff0000', '#ffffff'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'Elevation',
    bounds=imgs.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('JAXA_ALOS_AW3D30_V3_2_N027E118').select(['DSM'])

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=8
)

vis_params = {
    'bands': 'DSM',
    'min': 0,
    'max': 2000,
    'palette': [
        '#0000ff', '#00ffff', '#ffff00', '#ff0000', '#ffffff'
    ]
};

map.addLayer(
    img,
    vis_params,
    'Elevation',
    bounds=img.getBounds()
)
map
```
