## 公开矢量数据资源

公开矢量数据目前支持下列四类，类型为FeatureCollection，可以使用STAC ID导入。

|        STAC ID         |   说明    |
|:----------------------:|:-------:| 
|     China_Province     |  省份集合   |
|       China_City       |  城市集合   |
|     China_District     | 省/市/区集合 |
| WRI_GPPD_Power_Plants  | 全球发电站集合 |
