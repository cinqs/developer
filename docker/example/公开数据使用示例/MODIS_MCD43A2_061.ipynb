# MCD43A2 v061 简介与Notebook示例

MCD43A2 为BRDF/Albedo模型质量参数数据，为MCD43A3、MCD43A4数据产品提供了每日500米分辨率的BRDF/Albedo模型质量参数。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

dataset = aie.ImageCollection('MODIS_MCD43A2_061')\
             .filterDate('2022-11-01', '2022-12-01')\
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=2
)

vis_params = {
  'band':'Snow_BRDF_Albedo',
  'min': 0.0,
  'max': 1.0
}

map.addLayer(
    dataset.select(['Snow_BRDF_Albedo']), 
    vis_params, 
    'Snow_BRDF_Albedo',
     bounds=dataset.getBounds()
)

map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('MODIS_MCD43A2_061_2022_10_31')


map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=2
)


vis_params = {
  'band':'Snow_BRDF_Albedo',
  'min': 0.0,
  'max': 1.0
}

map.addLayer(
    img.select(['Snow_BRDF_Albedo']), 
    vis_params, 
    'Snow_BRDF_Albedo',
     bounds=img.getBounds()
)

map
```
