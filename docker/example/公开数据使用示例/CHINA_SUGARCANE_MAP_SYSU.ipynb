# 2016-2020年中国甘蔗种植分布数据集（袁文平教授团队）简介与Notebook示例
利用Landsat和Sentinel卫星影像，基于物侯识别和动态时间规划算法（TWDTW），生成了2016-2020年中国的甘蔗分布图。经验证，使用TOA生成的数据集的总体精度达到了92.74%。数据覆盖省份包括：广东省、广西省、海南省、云南省。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


dataset = aie.ImageCollection('CHINA_SUGARCANE_MAP_SYSU') \
             .limit(1)

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': ['Map'],
    'min': 0,
    'max': 1,
    "palette":["#d9d9d9","#ff0000"]
}

map.addLayer(
    dataset,
    vis_params,
    'CHINA_SUGARCANE_MAP_SYSU',
    bounds=dataset.getBounds()
)
map
```
