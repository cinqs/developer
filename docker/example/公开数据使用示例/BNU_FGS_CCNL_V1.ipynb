# 基于DMSP-OLS校正的1992-2013年夜光数据集（CCNL） 简介与Notebook示例

DMSP-OLS 提供了1992-2013年长时序的夜间灯光观测数据，第四版 DMSP-OLS夜光数据产品由NOAA负责校正和处理。本产品在第四版数据的基础上，使用时序分析方法来提升数据产品的精度，以此得到年际间一致性较好订正后的数据产品CCNL。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('BNU_FGS_CCNL_V1') \
             .filterDate("1992-01-01", "1992-01-31") \
             
imgs = dataset.select(['b1']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=4
)
vis_params = {
    'bands': ['b1'],
    'min': 0,
    'max': 20,
    "palette":["#060606","#f1f50b"]
}

map.addLayer(
    imgs,
    vis_params,
    'CCNL',
    bounds=imgs.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('BNU_FGS_CCNL_V1_DMSP1992V1')

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=4
)

vis_params = {
    'bands': ['b1'],
    'min': 0,
    'max': 20,
    "palette":["#060606","#f1f50b"]
}

map.addLayer(
    img,
    vis_params,
    'CCNL',
    bounds=img.getBounds()
)
map
```
