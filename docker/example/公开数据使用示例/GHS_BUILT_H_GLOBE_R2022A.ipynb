# GHS Building Height简介与Notebook示例
GHS Building Height 使用了AW3D30, SRTM30和Sentinel-2(2018年) 3套数据集生成，可用于表征建筑物高度。数据集时间为2018年、10om分辨率，单位为米。具体信息可查阅<a href="https://ghsl.jrc.ec.europa.eu/ghs_buH2022.php" target="_blank">https://ghsl.jrc.ec.europa.eu/ghs_buH2022.php</a> 。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定需要检索的区域
feature_collection = aie.FeatureCollection('China_Province') \
                        .filter(aie.Filter.eq('province', '浙江省'))

geometry = feature_collection.geometry()


dataset = aie.ImageCollection('GHS_BUILT_H_GLOBE_R2022A') \
             .filterBounds(geometry) \
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=4
)

vis_params = {
    'bands': ['AGBH'],
    'min': 0,
    'max': 100,
    "palette":["#1A1384","#BB5079","#F1F375"]
}

map.addLayer(
    dataset,
    vis_params,
    'GHS Building Height',
    bounds=dataset.getBounds()
)
map
```
