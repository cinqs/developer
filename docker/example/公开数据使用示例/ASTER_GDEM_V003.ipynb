# ASTGTM V003简介与Notebook示例
ASTER GDEM数据由1999年12月18日发射升空的Terra卫星搭载的ASTER仪器拍摄的ASTER Level 1A立体影像处理得到。
ASTER GDEM数据目前有三个版本，最新的版本为ASTER GDEM V003。相较于V002版本。ASTER GDEM V003采用的影像成像时间在2000/3/1~2013/11/30之间，空间分辨率为1弧度（30 m)。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定需要检索的区域
feature_collection = aie.FeatureCollection('China_Province') \
                        .filter(aie.Filter.eq('province', '浙江省'))

geometry = feature_collection.geometry()


dataset = aie.ImageCollection('ASTER_GDEM_V003') \
             .filterBounds(geometry) \
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=8
)

vis_params = {
    'bands': ['elevation'],
    'min': 0,
    'max': 2000,
}

map.addLayer(
    dataset,
    vis_params,
    'True Color',
    bounds=dataset.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('ASTER_GDEM_V003_S82E154')

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=8
)

vis_params = {
    'bands': ['elevation'],
    'min': 0,
    'max': 2000,
}

map.addLayer(
    img,
    vis_params,
    'True Color',
    bounds=img.getBounds()
)
map
```
