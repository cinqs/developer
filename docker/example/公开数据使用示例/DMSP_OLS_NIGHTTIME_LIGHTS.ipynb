# DMSP OLS Nighttime Lights Time Series V4简介与Notebook示例
该数据集为第四代DMSP-OLS夜间灯光数据（Version 4 DMSP-OLS Nighttime Lights Time Series）时间跨度为1992 年到2013 年，共计22年的影像时间序列。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


dataset = aie.ImageCollection('DMSP_OLS_NIGHTTIME_LIGHTS') \
             .filterDate('1993-01-01', '1994-12-31') \
             .limit(3);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=1
)

vis_params = {
    'bands': ['avg_vis'],
    'min': 0,
    'max': 20,
    "palette":["#060606","#f1f50b"]
}

map.addLayer(
    dataset,
    vis_params,
    'True Color',
    bounds=dataset.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('DMSP_OLS_NIGHTTIME_LIGHTS_F10_1992')

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=1
)

vis_params = {
    'bands': ['avg_vis'],
    'min': 0,
    'max': 20,
    "palette":["#060606","#f1f50b"]
}

map.addLayer(
    img,
    vis_params,
    'True Color',
    bounds=img.getBounds()
)
map
```
