# OpenLandMap Potential Distribution of Biomes 简介与Notebook示例

该数据集为潜在自然植被生物群落的全球类别预测（主要基于BIOMES 6000数据集的 "当前生物群落 "类别）
潜在自然植被（PNV）是指在某一特定地点不受人类活动影响而存在的与气候平衡的植被覆盖。PNV对于提高公众对土地退化的认识和估计土地潜力非常有用。这个数据集包含以下预测结果
(1) 基于BIOME 6000数据集( 8057个基于现代花粉的位点重建)的生物群落分布，
(2) 基于详细的发生记录（1,546,435次地面观测）的欧洲森林树种的分布，
(3) 全球每月吸收光合有效辐射的FAPAR值（包括30301个随机抽样的点）。

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.Image('OPENLANDMAP_PNV_PNV_BIOME-TYPE_BIOME00K_C_V01')
             
imgs = dataset.select(['biome_type']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=4
)
vis_params = {
    'bands': ['biome_type'],
    'min': 1.0,
    'max': 32.0,
    "palette":['#1c5510','#659208','#ae7d20','#000065','#bbcb35','#009a18',
            '#caffca','#55eb49','#65b2ff','#0020ca','#8ea228','#ff9adf',
            '#baff35','#ffba9a','#ffba35','#f7ffca','#e7e718','#798649',
            '#65ff9a','#d29e96']
}

map.addLayer(
    imgs.updateMask(imgs.gt(aie.Image(0))),
    vis_params,
    'OpenLandMap Potential Distribution of Biomes',
    bounds=imgs.getBounds()
)
map
```
