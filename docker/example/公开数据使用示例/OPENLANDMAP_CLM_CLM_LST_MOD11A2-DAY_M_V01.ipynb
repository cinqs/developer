# OpenLandMap Long-term Land Surface Temperature Daytime Monthly Median 简介与Notebook示例

该数据集为地表温度数据集，利用MOD11A2数据产品，计算2000-2017年月度均值。

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.Image('OPENLANDMAP_CLM_CLM_LST_MOD11A2-DAY_M_V01')
             
imgs = dataset.select(['jan']).multiply(aie.Image(0.02));

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=4
)
vis_params = {
    'bands': ['jan'],
    'min': 230,
    'max': 320,
    "palette":[
            '#0000ff', '#00ffff', '#ffff00', '#ff0000'
  ]
}

map.addLayer(
    imgs,
    vis_params,
    'OpenLandMap Long-term LST Daytime Monthly mean (K)',
    bounds=imgs.getBounds()
)
map
```
