# MOD11A2 V061 简介与Notebook示例

MOD11A2 V061 是基于Terra中分辨率成像分光仪（MODIS），提供的8天间隔、1 km分辨率的陆面温度和发射率（LST&E）产品。其中LST是8天内逐日相应的MOD11A1  LST的平均值。与白天黑夜 LST一起提供的还包括对应的质量控制、观测时间、观测天顶角、晴空覆盖率以及土地覆盖类型的31-32波段的辐射率等数据。
## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

dataset = aie.ImageCollection('MODIS_MOD11A2_061')\
             .filterDate('2022-11-01', '2022-11-30') 

imgs = dataset.select(['LST_Day_1km'])

map = aie.Map(
    center=[6.746, 46.529],
    height=800,
    zoom=1
)

vis_params = {
    'bands': 'LST_Day_1km',
    'min': 13000.0,
    'max': 16500.0,
    'palette': [
        '#040274', '#040281', '#0502a3', '#0502b8', '#0502ce', '#0502e6',
        '#0602ff', '#235cb1', '#307ef3', '#269db1', '#30c8e2', '#32d3ef',
        '#3be285', '#3ff38f', '#86e26f', '#3ae237', '#b5e22e', '#d6e21f',
        '#fff705', '#ffd611', '#ffb613', '#ff8b13', '#ff6e08', '#ff500d',
        '#ff0000', '#de0101', '#c21301', '#a71001', '#911003'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'Land Surface Temperature',
    bounds=imgs.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('MODIS_MOD11A2_061_2023_01_17').select(['LST_Day_1km'])


map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=2
)



vis_params = {
    'bands': 'LST_Day_1km',
    'min': 13000.0,
    'max': 16500.0,
    'palette': [
        '#040274', '#040281', '#0502a3', '#0502b8', '#0502ce', '#0502e6',
        '#0602ff', '#235cb1', '#307ef3', '#269db1', '#30c8e2', '#32d3ef',
        '#3be285', '#3ff38f', '#86e26f', '#3ae237', '#b5e22e', '#d6e21f',
        '#fff705', '#ffd611', '#ffb613', '#ff8b13', '#ff6e08', '#ff500d',
        '#ff0000', '#de0101', '#c21301', '#a71001', '#911003'
    ]
}

map.addLayer(
    img, 
    vis_params, 
    'True Color',
     bounds=img.getBounds()
)

map
```
