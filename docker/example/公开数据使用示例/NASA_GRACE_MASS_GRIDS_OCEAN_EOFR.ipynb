# GRACE Monthly Mass Grids - Ocean EOFR 简介与Notebook示例
GRCTellus（GRCTellus）月度质量网格海洋数据集。提供了相对2004-2010年平均基线的月度引力异常值。该数据集所包含的数据是以 "等水厚度"为单位。

该数据集是GRACE Tellus（GRCTellus）海洋数据集的过滤版本。EOFR 底压（OBP）网格是通过将常规GRC海洋网格产品的数据投射到海洋环流和潮汐模型（OMCT）的经验正交函数（EOFs）上获得的。这有效地过滤了GRACE数据中与OMCT海洋模型中的物理学和OBP变化不一致的信号。经EOFR过滤后重建的OBP场与雷达测高的海面高度一致性更好，减少冰原和冰川周围的泄漏，并降低了中纬度地区的噪声。

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('NASA_GRACE_MASS_GRIDS_OCEAN_EOFR')
             
imgs = dataset.select(['lwe_thickness_csr']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=1
)
vis_params = {
    'bands': ['lwe_thickness_csr'],
    'min': -10,
    'max': 10,
}

map.addLayer(
    imgs,
    vis_params,
    'lwe_thickness_csr',
    bounds=imgs.getBounds()
)
map
```
