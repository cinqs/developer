# OpenLandMap Clay Content 简介与Notebook示例

该数据集为在6个土壤标准深度下（0、10、30、60、100和200cm）预测的粘土含量（%，kg/kg）。数据空间分辨率为250米。

## 数据检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集
dataset = aie.Image('OPENLANDMAP_SOL_SOL_CLAY-WFRACTION_USDA-3A1A1A_M_V02')
                
imgs = dataset.select(['b0']);

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=2
)
vis_params = {
    'bands': 'b0',
    'min': 1.0,
    'max': 100.0,
    'palette': [
        "#FFFF00","#F8F806","#F1F10C","#EBEB13","#E4E419","#DDDD20",
        "#D7D726","#D0D02D","#CACA33","#C3C33A","#BCBC41","#B6B647",
        "#B0B04E","#A9A954","#A3A35A","#9C9C61","#959568","#8F8F6E",
        "#898975","#82827B","#7B7B82","#757589","#6E6E8F","#686895",
        "#61619C","#5A5AA3","#5454A9","#4D4DB0","#4747B6","#4141BC",
        "#3A3AC3","#3333CA","#2D2DD0","#2626D7","#2020DD","#1919E4",
        "#1212EB","#0C0CF1","#0606F8","#0000FF",
  ]
}
map.addLayer(
    imgs,
    vis_params,
    'Clay content',
    bounds=imgs.getBounds()
)
map
```
