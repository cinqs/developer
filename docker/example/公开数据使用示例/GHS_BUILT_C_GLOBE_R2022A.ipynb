# GHS Settlement Characteristics简介与Notebook示例
GHS Settlement Characteristics由Sentinel-2（2018年）和其他GHS R2022A数据衍生而来。数据集为10m分辨率的人类居住区边界数据，并从建筑环境的形态和功能使用描述其内部特征。
数据集的进一步分类是参考了植被强度（GHS-BUILT-C_VEG_GLOBE_R2022A）、水面（GHS_LAND_GLOBE_R2022A）、路面（OSM公路）、功能使用（GHS-BUILT-C_FUN_GLOBE_R2022A）和建筑高度（GHS-BUILT-H_GLOBE_R2022A）等数据集。具体信息可查阅
<a href="https://ghsl.jrc.ec.europa.eu/ghs_buC2022.php" target="_blank">https://ghsl.jrc.ec.europa.eu/ghs_buC2022.php</a>。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


dataset = aie.ImageCollection('GHS_BUILT_C_GLOBE_R2022A') \
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=4
)

vis_params = {
    'bands': ['MSZ'],
    'min': 1,
    'max': 255,
    "palette":["#76896F","#8AD66B","#CBFCA9","#52B2F5","#F7D548",
               "#C5852E","#E8652B","#E83424","#BB2827","#701212",
               "#EC71DC","#EC71DD","#E231F6","#9621F2","#640EF3"]
}

map.addLayer(
    dataset,
    vis_params,
    'GHS Settlement Characteristics',
    bounds=dataset.getBounds()
)
map
```
