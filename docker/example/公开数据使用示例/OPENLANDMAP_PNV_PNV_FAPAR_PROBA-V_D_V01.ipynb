# OpenLandMap Potential FAPAR Monthly 简介与Notebook示例

该数据集为潜在自然植被生物群落的月度吸收光合有效辐射的FAPAR值，基于PROB-V FAPAR 2014-2017数据集获得。

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.Image('OPENLANDMAP_PNV_PNV_FAPAR_PROBA-V_D_V01')
             
imgs = dataset.select(['jan']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=4
)
vis_params = {
    'bands': ['jan'],
    'min': 0.0,
    'max': 200.0,
    "palette":["#0000ff","#00ffff","#ffff00","#ff0000"]
}

map.addLayer(
    imgs,
    vis_params,
    'OpenLandMap Potential FAPAR Monthly',
    bounds=imgs.getBounds()
)
map
```
