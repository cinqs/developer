# GRACE Monthly Mass Grids - Land 简介与Notebook示例

GRCTellus（GRCTellus）月度质量网格海洋数据集。提供了相对2004-2010年平均基线的月度引力异常值。该数据集所包含的数据是以 "等水厚度 "为单位。<a href="https://grace.jpl.nasa.gov/data/monthly-mass-grids" target="_blank">【更多信息】</a>

GRACE Tellus（GRCTellus）月度质量网格海洋数据集CSR（德克萨斯大学/空间研究中心）、GFZ（波茨坦地质研究中心）和JPL（美国航空航天局喷气推进实验室）三个单位参与。每个中心都是GRACE地面系统的一部分，生产本数据集所使用的二级数据（球面谐波场）。输出包括重力场和用于计算纠偏场的球面谐波系数。由于每个中心独立产生系数，结果可能略有不同。对大多数用户来说，建议使用所有三个数据集的平均值。
<a href="https://grace.jpl.nasa.gov/data/choosing-a-solution/" target="_blank">【更多信息】</a>

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('NASA_GRACE_MASS_GRIDS_LAND')
             
imgs = dataset.select(['lwe_thickness_csr']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=1
)
vis_params = {
    'bands': ['lwe_thickness_csr'],
    'min': -25,
    'max': 25,
    "palette":[
             '#ffffff','#d0d0d0','#ababab','#ffb93d','#3977ff'
  ]
}

map.addLayer(
    imgs,
    vis_params,
    'lwe_thickness',
    bounds=imgs.getBounds()
)
map
```
