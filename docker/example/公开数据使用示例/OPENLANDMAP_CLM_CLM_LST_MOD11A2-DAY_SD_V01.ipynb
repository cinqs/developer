# OpenLandMap Long-term Land Surface Temperature Daytime Monthly Standard Deviation 简介与Notebook示例

该数据集为地表温度标准偏差数据集。由MOD1A2数据产品衍生而来，计算了MODIS LST数据昼夜温度的标准偏差。

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.Image('OPENLANDMAP_CLM_CLM_LST_MOD11A2-DAY_SD_V01')
             
imgs = dataset.select(['jan']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=2
)
vis_params = {
    'bands': ['jan'],
    'min': 0,
    'max': 380,
    "palette":[
            '#2828ff','#2828ff','#6666ff','#8989ff','#a1a1ff','#b2b2ff',
            '#c0c0ff','#cbcbff','#d5d5ff','#dedeff','#e6e6ff','#ededff',
            '#f5f5ff','#ffffff','#fcfcff','#fffbfb','#fff4f4','#ffeded',
            '#ffe5e5','#ffdddd','#ffd4d4','#ffcbcb','#ffbfbf','#ffb2b2',
            '#ffa1a1','#ff8a8a','#ff6767','#ff2929',
  ]
}

map.addLayer(
    imgs,
    vis_params,
    'OpenLandMap Long-term LST Daytime Monthly SD',
    bounds=imgs.getBounds()
)
map
```
