# MCD43A3 v061 简介与Notebook示例

MCD43A3 反照率模型数据集逐日的16天数据产品。它提供了MODIS表面反射波段（band 1-7）以及可见光、近红外和短波的定向半球反射率（黑色天空反照率）和双半球反射率（白色天空反照率）。500米分辨率的逐日影像是经16天的数据生成的，以给定日期中心。并为10个波段均提供对应的质量波段。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

dataset = aie.ImageCollection('MODIS_MCD43A3_061')\
             .filterDate('2022-11-01', '2022-12-01')\
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=2
)


vis_params = {
  'band':'Albedo_BSA_Band1',
  'min': 0.0,
  'max': 400.0
}

map.addLayer(
    dataset.select(['Albedo_BSA_Band1']), 
    vis_params, 
    'Albedo_BSA_Band1',
     bounds=dataset.getBounds()
)

map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('MODIS_MCD43A3_061_2022_10_31')


map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=2
)

vis_params = {
  'band':'Albedo_BSA_Band1',
  'min': 0.0,
  'max': 400.0
}

map.addLayer(
    img.select(['Albedo_BSA_Band1']), 
    vis_params, 
    'Albedo_BSA_Band1',
     bounds=img.getBounds()
)

map
```
