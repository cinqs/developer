# SRTMGL1 V003简介与Notebook示例
SRTM（The Shuttle Radar Topography Mission）为美国航天航空局（NASA）、美国国家地理空间情报局（NGA）和德国航空太空中心（DLR）合作项目，获取60°N到56°S所有陆域的数字高程模型。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定需要检索的区域
feature_collection = aie.FeatureCollection('China_Province') \
                        .filter(aie.Filter.eq('province', '浙江省'))

geometry = feature_collection.geometry()

dataset = aie.ImageCollection('SRTM90_V003') \
             .filterBounds(geometry) \
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': ['elevation'],
    'min': 0,
    'max': 2000,
}

map.addLayer(
    dataset,
    vis_params,
    'True Color',
    bounds=dataset.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('SRTM90_V003_N00E006')

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=8
)

vis_params = {
    'bands': ['elevation'],
    'min': 0,
    'max': 2000,
}

map.addLayer(
    img,
    vis_params,
    'True Color',
    bounds=img.getBounds()
)
map
```
