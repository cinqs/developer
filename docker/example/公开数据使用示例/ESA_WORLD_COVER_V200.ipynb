# ESA WorldCover 10m v200 简介与Notebook示例
欧洲航天局 (ESA) WorldCover 10m土地覆盖数据集是基于Sentinel-1和 Sentinel-2 数据制作的2021年覆盖全球的土地覆盖分类数据产品，共有11个土地覆盖类型分类，分别是林地、灌木地、草地、耕地、人造地表、裸地/稀疏植被、冰川/积雪、水体、湿地、红树林、苔原。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定需要检索的区域
feature_collection = aie.FeatureCollection('China_Province') 

#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('ESA_WORLD_COVER_V200') \
             .filterBounds(feature_collection) \
             .filterDate("2020-01-01", "2022-01-31")

imgs = dataset.select(['Map']);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=4
)

vis_params = {
    'bands': 'Map',
    'min': 10.0,
    'max': 100.0,
    'palette': [
        '#006400','#ffbb22','#ffff4c','#f096ff',
        '#fa0000','#b4b4b4','#f0f0f0','#0064c8',
        '#0096a0','#00cf75','#fae6a0'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'Map',
    bounds=imgs.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('ESA_WorldCover_10m_2021_v200_N30E120').select(['Map'])

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': 'Map',
    'min': 10.0,
    'max': 100.0,
    'palette': [
        '#006400','#ffbb22','#ffff4c','#f096ff',
        '#fa0000','#b4b4b4','#f0f0f0','#0064c8',
        '#0096a0','#00cf75','#fae6a0'
    ]
}

map.addLayer(
    img,
    vis_params,
    'Map',
    bounds=img.getBounds()
)
map
```
