# OpenLandMap Precipitation Monthly 简介与Notebook示例

该数据集是基于SM2RAIN-ASCAT 2007-2018, IMERG, CHELSA Climate, and WorldClim等数据，进行融合与降尺度计算得到的1km分辨率的全球月度降水量。请注意该数据不覆盖南极洲。

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.Image('OPENLANDMAP_CLM_CLM_PRECIPITATION_SM2RAIN_M_V01')
             
imgs = dataset.select(['jan']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=2
)
vis_params = {
    'bands': ['jan'],
    'min': 0,
    'max': 380,
    "palette":['#ecffbd', '#ffff00', '#3af6ff', '#467aff', '#313eff', '#0008ff']
}

map.addLayer(
    imgs,
    vis_params,
    'OpenLandMap Precipitation Monthly',
    bounds=imgs.getBounds()
)
map
```
