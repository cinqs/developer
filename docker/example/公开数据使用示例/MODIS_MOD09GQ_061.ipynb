# MOD09GQ v061 简介与Notebook示例

MOD09GQ v061 提供了Terra中分辨率250米1-2波段表面光谱反射率的估算产品，并进行了大气校正（如气体、气溶胶和瑞利散射）。同时提供与反射率产品对应的质量（QA）层和五个观测层。该产品可与500米（MOD09GA）产品的质量和观测信息一起使用。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

dataset = aie.ImageCollection('MODIS_MOD09GQ_061')\
             .filterDate('2022-11-01', '2022-12-01')\
             .limit(10);
imgs = dataset.select(['sur_refl_b02', 'sur_refl_b02', 'sur_refl_b01'])

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=2
)


vis_params = {
    'bands': ['sur_refl_b02', 'sur_refl_b02', 'sur_refl_b01'],
    'min': -100,
    'max': 8000.0
}

map.addLayer(
    imgs, 
    vis_params, 
    'True Color',
     bounds=imgs.getBounds()
)

map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('MODIS_MOD09GQ_061_2022_12_02').select(['sur_refl_b02', 'sur_refl_b02', 'sur_refl_b01'])


map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=2
)


vis_params = {
    'bands': ['sur_refl_b02', 'sur_refl_b02', 'sur_refl_b01'],
    'min': -100,
    'max': 8000.0
}

map.addLayer(
    img, 
    vis_params, 
    'True Color',
     bounds=img.getBounds()
)

map
```
