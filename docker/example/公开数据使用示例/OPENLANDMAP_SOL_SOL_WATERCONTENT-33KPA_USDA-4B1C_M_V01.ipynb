# OpenLandMap Soil Water Content at 33kPa (Field Capacity) 简介与Notebook示例

该数据集为在6个土壤标准深度下（0、10、30、60、100和200cm）预测的33kPa吸力对应的土壤体积含水量（%）。

## 数据检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集
dataset = aie.Image('OPENLANDMAP_SOL_SOL_WATERCONTENT-33KPA_USDA-4B1C_M_V01')
                
imgs = dataset.select(['b0']);

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=2
)
vis_params = {
    'bands': 'b0',
    'min': 0,
    'max': 52.0,
    'palette': [
       "#d29642","#eec764","#b4ee87","#32eeeb","#0c78ee","#2601b7",
    "#083371"
    ]
}
map.addLayer(
    imgs,
    vis_params,
    'Soil water content at 33kPa (field capacity)',
    bounds=imgs.getBounds()
)
map
```
