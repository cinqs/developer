# OpenLandMap Predicted Hapludalfs Probability 简介与Notebook示例

该数据集为USDA预测的在250米深度的土壤分类概率产品。数据集根据全球土壤剖面汇编数据经机器学习方法进行预测生成。

## 数据检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集
dataset = aie.Image('OPENLANDMAP_SOL_SOL_GRTGROUP_USDA-SOILTAX-HAPLUDALFS_P_V01')
                
imgs = dataset.select(['grtgroup']);

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=2
)
vis_params = {
    'bands': 'grtgroup',
    'min': 0.0,
    'max': 35.0,
    'palette': [
        '#ffffb2', '#fecc5c', '#fd8d3c', '#f03b20', '#bd0026'
    ]
}
map.addLayer(
    imgs,
    vis_params,
    'Soil Predicted Hapludalfs Probability',
    bounds=imgs.getBounds()
)
map
```
