# MCD43A4 v061 简介与Notebook示例

MCD43A4 数据为Modis数据提供了1-7波段的NBAR(Nadir Bidirectional Reflectance Distribution Function (BRDF)-Adjusted Reflectance)和简化的强制质量层。在使用该产品是，应当相应的考虑MCD43 A2数据集中的质量信息。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

dataset = aie.ImageCollection('MODIS_MCD43A4_061')\
             .filterDate('2022-11-01', '2022-12-01')\
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=2
)

vis_params = {
    'bands': ['Nadir_Reflectance_Band1', 'Nadir_Reflectance_Band4' , 'Nadir_Reflectance_Band3'],
    'min': 0.0,
    'max': 4000.0
}

map.addLayer(
    dataset, 
    vis_params, 
    'map visualization',
     bounds=dataset.getBounds()
)

map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('MODIS_MCD43A4_061_2022_10_31')

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=2
)


vis_params = {
    'bands': ['Nadir_Reflectance_Band1', 'Nadir_Reflectance_Band4','Nadir_Reflectance_Band3'],
    'min': 0.0,
    'max': 4000.0
}

map.addLayer(
    img, 
    vis_params, 
    'map visualization',
     bounds=img.getBounds()
)

map
```
