# MCD43A1 v061 简介与Notebook示例

MCD43A1 为BRDF/Albedo模型参数数据产品，利用了16天500米分辨率的 MODIS Terra 和 Aqua数据制作而成。提供了每日500米分辨率的BRDF/Albedo模型参数数据集。
MCD43A1为MODIS1-7光谱段及可见光、近红外（NIR）和短波带提供了三个模型加权参数。与这些波段的三维参数层一起的是10个波段中每个波段的简化强制质量层。在使用该产品时，应参考相应的MCD43A2数据文件中提供的基本质量信息。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

dataset = aie.ImageCollection('MODIS_MCD43A1_061')\
             .filterDate('2022-11-01', '2022-12-01')\
             .limit(10);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=2
)


vis_params = {
    'bands': ['BRDF_Albedo_Parameters_Band1_iso', 'BRDF_Albedo_Parameters_Band4_iso','BRDF_Albedo_Parameters_Band3_iso'],
    'min': 0.0,
    'max': 1400.0
}

map.addLayer(
    dataset, 
    vis_params, 
    'map visualization',
     bounds=dataset.getBounds()
)

map
```
