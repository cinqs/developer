# MOD09GA v061 简介与Notebook示例

MOD09GA v061 提供了Terra 500米中分辨率1-7波段表面光谱反射率的估算产品，并进行了大气校正（如气体、气溶胶和瑞利散射）。同时提供与500米反射率产品对应的1000米级别的观测和QA波段以及地理定标数据。MOD09GA不同波段反射率数据是其他Modis陆地产品的源数据。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

dataset = aie.ImageCollection('MODIS_MOD09GA_061')\
             .filterDate('2022-11-01', '2022-11-02')\
             .limit(10);
imgs = dataset.select(['sur_refl_b01', 'sur_refl_b04', 'sur_refl_b03'])

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=2
)


vis_params = {
    'bands': ['sur_refl_b01', 'sur_refl_b04', 'sur_refl_b03'],
    'min': -100,
    'max': 8000.0
}

map.addLayer(
    imgs, 
    vis_params, 
    'True Color',
     bounds=imgs.getBounds()
)

map
```
