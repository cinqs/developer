# OpenLandMap Soil Bulk Density 简介与Notebook示例

该数据集为在6个土壤标准深度下（0、10、30、60、100和200cm）预测的土壤容重（10 x kg/m3）。数据空间分辨率为250米。

## 数据检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集
dataset = aie.Image('OPENLANDMAP_SOL_SOL_BULKDENS-FINEEARTH_USDA-4A1H_M_V02')
                
imgs = dataset.select(['b0']);

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=2
)
vis_params = {
    'bands': 'b0',
    'min': 5.0,
    'max': 185.0,
    'palette': [
        '#5e3c99', '#b2abd2', '#f7e0b2', '#fdb863', '#e63b01'
  ]
}
map.addLayer(
    imgs,
    vis_params,
    'Soil bulk density',
    bounds=imgs.getBounds()
)
map
```
