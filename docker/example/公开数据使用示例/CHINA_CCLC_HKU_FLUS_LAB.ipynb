# 中国10米海岸带土地覆盖分类数据集(CCLC) 简介与Notebook示例

该数据产品精细刻画了中国海岸带土地覆盖空间格局，兼顾海域和陆域关键地物类型。数据生产采用2019~2021年哨兵主被动遥感数据构建光谱特征和纹理特征，联合基于像素和面向对象的分类方法，构建了2020年10米中国海岸带土地覆盖数据集，制图结果的总体分类精度达到79.3%。该研究提出的采样策略和制图框架可以推广到大尺度海岸带遥感制图中，数据产品将为海岸带环境遥感评估和监测提供重要依据。



## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定需要检索的区域
feature_collection = aie.FeatureCollection('China_Province') \
                       
#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('CHINA_CCLC_HKU_FLUS_LAB') \
             .filterBounds(feature_collection) \
             .filterDate("2020-01-01", "2020-01-31") \
             .min()
imgs = dataset.select(['Map']);

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=4
)
vis_params = {
    'bands': 'Map',
    'min': 1,
    'max': 17,
    'palette': [
       '#FA0000','#FA0001','#FA0002','#FA0003','#F095FE','#F095FF','#006407',
        '#FFBB22','#F9E59F','#00CF75',
        '#0296A0','#A7A70E','#A7A70F','#0064C8','#B4B4B4','#E69804','#FFFF4C'
    ]
}
map.addLayer(
    imgs,
    vis_params,
    'Map',
    bounds=imgs.getBounds()
)
map
```
