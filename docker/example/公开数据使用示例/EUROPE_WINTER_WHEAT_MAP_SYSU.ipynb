# 2016-2020年欧洲冬季谷物种植分布数据集（袁文平教授团队）简介与Notebook示例
使用Sentinel-1 A / B的合成孔径雷达数据，区分油菜籽和冬季谷物；采用基于时间加权的动态时间规划算法（TWDTW），通过比较Landsat和Sentinel-2植被指数（NDVI）中季节变化的相似性，把冬季谷物与其他作物区分开来。在此基础上生成了 2016-2020 年的欧洲冬季谷物分布图，覆盖 32个欧洲国家，空间分辨率为30米。验证表明，数据集的生产者和用户平均精度分别为91%±7.8%和89%±10.3%。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


dataset = aie.ImageCollection('EUROPE_WINTER_WHEAT_MAP_SYSU') \
             .limit(1)

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': ['Map'],
    'min': 0,
    'max': 1,
    "palette":["#d9d9d9","#ff0000"]
}

map.addLayer(
    dataset,
    vis_params,
    'EUROPE_WINTER_WHEAT_MAP_SYSU',
    bounds=dataset.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('EUROPE_WINTER_WHEAT_MAP_SYSU_2020_N45E21')

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': ['Map'],
    'min': 0,
    'max': 1,
    "palette":["#d9d9d9","#ff0000"]
}

map.addLayer(
    img,
    vis_params,
    'EUROPE_WINTER_WHEAT_MAP_SYSU',
    bounds=img.getBounds()
)
map
```
