# NASA_GRACE_MASS_GRIDS_MASCON_CRI 简介与Notebook示例
GRCTellus（GRCTellus）月度质量网格海洋数据集。提供了相对2004-2010年平均基线的月度引力异常值。该数据集所包含的数据是以 "等水厚度 "为单位。<a href="https://grace.jpl.nasa.gov/data/monthly-mass-grids" target="_blank">【更多信息】</a>


GRACE Tellus（GRCTellus）月度质量数据集由JPL基于GRACE观测Level -1 数据生产而来。该数据集利用空间和时间上的先验约束，以等面积3°x3°球冠质量浓度（mascon）函数来估计全球月度重力场，以减少测量误差影响。与传统的球面谐波解决方案相比，这使得mascon场的信噪比更好。

这个版本的全球Mascon数据集有一个海岸线分辨率改进（CRI）过滤器，用于在后处理步骤中分离每个陆地/海洋mascon中的陆地和海洋部分的质量。对于希望自己分离陆地和海洋质量信号的用户，可以在NASA/GRACE/MASS_GRIDS/MASCON上找到没有CRI滤波器的数据版本。

# 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('NASA_GRACE_MASS_GRIDS_MASCON_CRI')
             
imgs = dataset.select(['lwe_thickness']);

map = aie.Map(
     center=[121.02,31.17],
    height=800,
    zoom=4
)
vis_params = {
    'bands': ['lwe_thickness'],
    'min': -25,
    'max': 25,
    "palette":[
             '#ffffff','#d0d0d0','#ababab','#ffb93d','#3977ff'
  ]
}

map.addLayer(
    imgs,
    vis_params,
    'OpenLandMap Long-term LST Monthly Day-Night Difference',
    bounds=imgs.getBounds()
)
map
```
