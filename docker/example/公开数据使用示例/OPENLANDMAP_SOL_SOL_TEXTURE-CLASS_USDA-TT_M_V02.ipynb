# OpenLandMap Soil Texture Class 简介与Notebook示例

该数据集为在6个土壤标准深度下（0、10、30、60、100和200cm）预测的土壤质地分类。数据空间分辨率为250米。

## 数据检索

```python
import aie
aie.Authenticate()
aie.Initialize()


#指定检索数据集
dataset = aie.Image('OPENLANDMAP_SOL_SOL_TEXTURE-CLASS_USDA-TT_M_V02')
                
imgs = dataset.select(['b0']);

map = aie.Map(
    center=imgs.getCenter(),
    height=800,
    zoom=2
)
vis_params = {
    'bands': 'b0',
    'min': 1.0,
    'max': 12.0,
    'palette': [
        "#d5c36b","#b96947","#9d3706","#ae868f","#f86714","#46d143",
        "#368f20","#3e5a14","#ffd557","#fff72e","#ff5a9d","#ff005b",
  ]
}
map.addLayer(
    imgs,
    vis_params,
    'Soil texture class (USDA system)',
    bounds=imgs.getBounds()
)
map
```
