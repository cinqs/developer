# 2016-2020年中国玉米种植分布数据集（袁文平教授团队）简介与Notebook示例
利用Landsat和Sentinel卫星获取的植被指数影像，通过基于时间权重的动态时间规划算法（TWDTW），生成了2016–2020年全国22个省份的30m分辨率玉米种植图，平均生产者精度与用户精度分别达到76.15%和81.59%。数据覆盖省份包括：安徽省、重庆市、甘肃省、广西省、贵州省、河北省、河北省、黑龙江省、河南省、湖北省、湖南省、内蒙古自治区、江苏省、吉林省、辽宁省、宁夏回族自治区、山西省、陕西省、四川省、天津市、新疆维吾尔自治区、云南省。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()


dataset = aie.ImageCollection('CHINA_MAIZE_MAP_SYSU') \
             .limit(1)

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': ['Map'],
    'min': 0,
    'max': 1,
    "palette":["#d9d9d9","#ff0000"]
}

map.addLayer(
    dataset,
    vis_params,
    'CHINA_MAIZE_MAP_SYSU',
    bounds=dataset.getBounds()
)
map
```
