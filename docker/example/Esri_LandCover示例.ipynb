# ESRI地物分类 简介与Notebook示例
ESRI 10m 地物覆盖数据产品是基于欧洲航天局(European Space Agency)10米分辨率Sentinel-2 影像数据，Impact Observatory 提供地物分类AI模型、微软提供计算资源，与Esri联合推出的10m地物分类数据产品。共有10种土地覆盖类型分类，分别是水体、林地、湿地、耕地、人造地表、裸地、冰川/积雪、云、草地。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定需要检索的区域
feature_collection = aie.FeatureCollection('China_Province') \
                        .filter(aie.Filter.eq('province', '浙江省'))

geometry = feature_collection.geometry()


#指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('ESRI_SENTINEL2_10M_LC') \
             .filterBounds(geometry) \
             .filterDate("2020-01-01", "2022-01-31") \
             .limit(100)
imgs = dataset.select(['Map']);

map = aie.Map(
    center=dataset.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': 'Map',
    'min': 1.0,
    'max': 11.0,
    'palette': [
        '#1e508f','#326f23','#74af85',
        '#d3b652','#c6162b','#c4c2bd',
        '#c9cfd2','#a7a7a7','#c7ad8e'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'Map',
    bounds=imgs.getBounds()
)
map
```

## 单景影像检索

```python
import aie
aie.Authenticate()
aie.Initialize()

img = aie.Image('ESRI_LC_51R_20210101_20220101').select(['Map'])

map = aie.Map(
    center=img.getCenter(),
    height=800,
    zoom=6
)

vis_params = {
    'bands': 'Map',
    'min': 1.0,
    'max': 11.0,
    'palette': [
        '#1e508f','#326f23','#74af85',
        '#d3b652','#c6162b','#c4c2bd',
        '#c9cfd2','#a7a7a7','#c7ad8e'
    ]
}

map.addLayer(
    img,
    vis_params,
    'Map',
    bounds=img.getBounds()
)
map
```
