# MOD17A2H v006简介与Notebook示例

8天合成的500米分辨率总初级生产力（GPP）数据产品（L4），可用作表面光和作用、蒸发蒸腾作用、以及初级净第一生产力等模型的输入数据。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('MODIS_MOD17A2H_006') \
             .filterDate('2021-05-01', '2021-05-31') \
             .limit(10)

imgs = dataset.select(['Gpp'])

map = aie.Map(
    center=[6.746, 46.529],
    height=800,
    zoom=1
)

vis_params = {
    'bands': 'Gpp',
    'min': 0,
    'max': 600,
    'palette': [
        '#bbe029', '#0a9501', '#074b03'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'Gpp',
    bounds=imgs.getBounds()
)
map
```
