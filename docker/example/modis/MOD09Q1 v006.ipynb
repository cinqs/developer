# MOD09Q1 v006简介与Notebook示例

提供250米分辨率下波段1和2的地表光谱反射率估计值，是基于MOD09GQ的8天合成三级数据产品（L3），依据高观测覆盖率、低视角、无云或云阴影、气溶胶负载等条件确定每个像素值，并针对气体、气溶胶和瑞利散射等大气条件进行了校正。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('MODIS_MOD09Q1_006') \
             .filterDate('2021-05-01', '2021-05-31') \
             .limit(10)

imgs = dataset.select(['sur_refl_b01'])

map = aie.Map(
    center=[6.746, 46.529],
    height=800,
    zoom=1
)

vis_params = {
    'bands': 'sur_refl_b01',
    'min': -100,
    'max': 16000,
}

map.addLayer(
    imgs,
    vis_params,
    'sur_refl_b01',
    bounds=imgs.getBounds()
)
map
```
