# MOD17A3HGF v006简介与Notebook示例

年度全球500米分辨率净初级生产力（NPP）数据产品（L4），NPP作为陆地生态系统碳循环的一个重要组成成分，是评价生态结构功能稳定性的重要指标之一。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('MODIS_MOD17A3HGF_006') \
             .filterDate('2021-05-01', '2021-05-31') \
             .limit(10)

imgs = dataset.select(['Npp'])

map = aie.Map(
    center=[6.746, 46.529],
    height=800,
    zoom=1
)

vis_params = {
    'bands': 'Npp',
    'min': 0,
    'max': 19000,
    'palette': [ 
        '#bbe029', '#0a9501', '#074b03'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'Npp',
    bounds=imgs.getBounds()
)
map
```
