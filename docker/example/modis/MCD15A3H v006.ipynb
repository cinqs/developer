# MCD15A3H v006简介与Notebook示例

全球4天合成LAI产品数据集（L4），分辨率为500米，根据MCDIS叶面积指数（LAI）和光合有效辐射分数（FPAR）的4天合成产品中选取其中FPAR最大的LAI值作为4天合成的LAI值。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('MODIS_MCD15A3H_006') \
             .filterDate('2021-05-01', '2021-05-31') \
             .limit(10)

imgs = dataset.select(['Fpar'])

map = aie.Map(
    center=[6.746, 46.529],
    height=800,
    zoom=1
)

vis_params = {
    'bands': 'Fpar',
    'min': 0,
    'max': 100,
    'palette': [
        '#e1e4b4', '#999d60', '#2ec409', '#0a4b06'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'Fpar',
    bounds=imgs.getBounds()
)
map
```
