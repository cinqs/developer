# MCD64A1 v006简介与Notebook示例

每月一次的全球网格化500米产品（L3），包含每像素的燃烧区和质量信息。MCD64A1烧毁面积绘图方法采用了500米MODIS表面反射图像和1公里MODIS主动火灾观测。

## 数据集检索

```python
import aie
aie.Authenticate()
aie.Initialize()

# 指定检索数据集，可设置检索的时间范围
dataset = aie.ImageCollection('MODIS_MCD64A1_006') \
             .filterDate('2021-05-01', '2021-05-31') \
             .limit(10)

imgs = dataset.select(['BurnDate'])

map = aie.Map(
    center=[6.746, 46.529],
    height=800,
    zoom=1
)

vis_params = {
    'bands': 'BurnDate',
    'min': 30,
    'max': 341,
    'palette': [ 
        '#4e0400', '#951003', '#c61503', '#ff1901'
    ]
}

map.addLayer(
    imgs,
    vis_params,
    'BurnDate',
    bounds=imgs.getBounds()
)
map
```
